```python
from langchain_postgres import PostgresChatMessageHistory
from langchain_core.messages import HumanMessage, AIMessage
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_community.llms import Ollama
from langchain_community.embeddings import OllamaEmbeddings
from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferMemory
import psycopg
import uuid

# Database connection string (update with your cloud PostgreSQL credentials)
conn_info = "postgresql://postgres:mypassword@localhost:5432/mydb"  # Replace with cloud DB credentials in Colab

# Create a synchronous connection
try:
    sync_connection = psycopg.connect(conn_info)
except Exception as e:
    print(f"Failed to connect to PostgreSQL: {e}")
    raise

# Create table if it doesn't exist
try:
    PostgresChatMessageHistory.create_tables(sync_connection, "message_store")
except Exception as e:
    print(f"Error creating table: {e}")
    sync_connection.close()
    raise

# Generate a valid UUID for session_id
session_id = str(uuid.uuid4())

# Initialize chat history
chat_history = PostgresChatMessageHistory(
    "message_store",
    session_id,
    sync_connection=sync_connection
)

# Initialize ConversationBufferMemory with PostgresChatMessageHistory
memory = ConversationBufferMemory(
    chat_memory=chat_history,
    return_messages=True
)

# Initialize Ollama LLM (Phi-4) and embeddings
# Note: In Colab, Ollama must be accessed via a remote server or API
llm = Ollama(model="phi4")
embeddings = OllamaEmbeddings(model="phi4")

# Define the prompt template
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful AI assistant. Provide concise and accurate responses."),
    MessagesPlaceholder(variable_name="history"),
    ("human", "{input}")
])

# Create conversation chain
conversation = ConversationChain(
    llm=llm,
    prompt=prompt,
    memory=memory,
    verbose=False
)

# Function to display chat history from PostgreSQL
def display_chat_history(session_id):
    try:
        cursor = sync_connection.cursor()
        cursor.execute(f"SELECT session_id, message, created_at FROM message_store WHERE session_id = %s ORDER BY created_at", (session_id,))
        rows = cursor.fetchall()
        print("\n=== Chat History Table ===")
        print("Session ID | Message | Created At")
        print("-" * 80)
        for row in rows:
            session_id, message, created_at = row
            print(f"{session_id} | {message.get('data')} ({message.get('type')}) | {created_at}")
        cursor.close()
    except Exception as e:
        print(f"Error displaying chat history: {e}")

try:
    # Example interaction
    user_input = "Hello! Can you tell me about the weather?"
    response = conversation.invoke({"input": user_input})
    print(f"User: {user_input}")
    print(f"AI: {response['response']}")

    # Add another message
    user_input = "What's the capital of France?"
    response = conversation.invoke({"input": user_input})
    print(f"User: {user_input}")
    print(f"AI: {response['response']}")

    # Display the chat history from the database
    display_chat_history(session_id)

finally:
    # Close the connection
    try:
        sync_connection.close()
    except:
        pass
```

```python
from langchain_core.messages import HumanMessage, AIMessage, BaseMessage, message_to_dict, messages_from_dict
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_community.llms import Ollama
from langchain_community.embeddings import OllamaEmbeddings
from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferMemory
from pymongo import MongoClient
from datetime import datetime
import uuid
from typing import List
from langchain_core.chat_history import BaseChatMessageHistory

# Custom MongoDB Chat Message History
class MongoDBChatMessageHistory(BaseChatMessageHistory):
    def __init__(self, collection, session_id: str):
        self.collection = collection
        self.session_id = session_id
        self.messages: List[BaseMessage] = self.get_messages() # Add this line to initialize messages

    def add_message(self, message: BaseMessage) -> None:
        """Add a message to the MongoDB collection."""
        message_dict = message_to_dict(message)
        message_dict["session_id"] = self.session_id
        message_dict["created_at"] = datetime.utcnow()
        self.collection.insert_one(message_dict)
        self.messages.append(message) # Add this line to update the in-memory list

    def get_messages(self) -> List[BaseMessage]:
        """Retrieve all messages for the session, sorted by creation time."""
        cursor = self.collection.find({"session_id": self.session_id}).sort("created_at", 1)
        messages = [messages_from_dict([doc])[0] for doc in cursor]
        return messages

    def clear(self) -> None:
        """Clear all messages for the session."""
        self.collection.delete_many({"session_id": self.session_id})
        self.messages = [] # Add this line to clear the in-memory list

    @staticmethod
    def create_collection(collection):
        """Ensure the collection exists (MongoDB creates it automatically on first insert)."""
        collection.create_index("session_id")

# Database connection string (update with your MongoDB credentials)
mongo_uri = "mongodb://localhost:27017/"  # Replace with your MongoDB URI
db_name = "mydb"
collection_name = "message_store"

# Create a synchronous connection
try:
    client = MongoClient(mongo_uri)
    db = client[db_name]
    collection = db[collection_name]
except Exception as e:
    print(f"Failed to connect to MongoDB: {e}")
    raise

# Ensure collection is ready (create index)
try:
    MongoDBChatMessageHistory.create_collection(collection)
except Exception as e:
    print(f"Error setting up collection: {e}")
    client.close()
    raise

# Generate a valid UUID for session_id
session_id = str(uuid.uuid4())

# Initialize chat history
chat_history = MongoDBChatMessageHistory(collection=collection, session_id=session_id)

# Initialize ConversationBufferMemory with MongoDBChatMessageHistory
memory = ConversationBufferMemory(
    chat_memory=chat_history,
    return_messages=True
)

# Initialize Ollama LLM (Phi-4) and embeddings
llm = Ollama(model="phi4")
embeddings = OllamaEmbeddings(model="phi4")

# Define the prompt template
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful AI assistant. Provide concise and accurate responses."),
    MessagesPlaceholder(variable_name="history"),
    ("human", "{input}")
])

# Create conversation chain
conversation = ConversationChain(
    llm=llm,
    prompt=prompt,
    memory=memory,
    verbose=False
)

# Function to display chat history from MongoDB
def display_chat_history(session_id):
    try:
        cursor = collection.find({"session_id": session_id}).sort("created_at", 1)
        print("\n=== Chat History Collection ===")
        print("Session ID | Message | Created At")
        print("-" * 80)
        for doc in cursor:
            session_id = doc["session_id"]
            message_data = doc.get("data", "")
            message_type = doc.get("type", "")
            created_at = doc.get("created_at", "")
            print(f"{session_id} | {message_data} ({message_type}) | {created_at}")
    except Exception as e:
        print(f"Error displaying chat history: {e}")

try:
    # Example interaction
    user_input = "Hello! Can you tell me about the weather?"
    response = conversation.invoke({"input": user_input})
    print(f"User: {user_input}")
    print(f"AI: {response['response']}")

    # Add another message
    user_input = "What's the capital of France?"
    response = conversation.invoke({"input": user_input})
    print(f"User: {user_input}")
    print(f"AI: {response['response']}")

    # Display the chat history from the database
    display_chat_history(session_id)

finally:
    # Close the connection
    try:
        client.close()
    except:
        pass
```
